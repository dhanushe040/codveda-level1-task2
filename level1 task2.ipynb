```python
df=pd.read_csv(r'D:\codveda\Data Set For Task\Churn Prdiction Data\churn-bigml-80.csv')

selected_states = ['CA', 'NY', 'TX','AZ','VA','NE','WY']
filtered_df = df[df['State'].isin(selected_states)]
# Create histogram
counts, bins, patches = plt.hist(filtered_df['State'], bins=len(selected_states), edgecolor='black', color='skyblue')
# Add value labels using bar_label
plt.bar_label(patches, labels=[f'{int(c)}' for c in counts], label_type='edge', fontsize=10)
# Add axis labels and title
plt.xlabel('State')
plt.ylabel('Number of Customers')
plt.title('Customer Distribution Across Selected States')
plt.show()
```

```python
# Group by State and calculate average
grouped = df.groupby('State')['Total day calls'].mean().reset_index()

# Scatter plot
plt.figure(figsize=(12,6))
plt.scatter(grouped['State'], grouped['Total day calls'], color='red')
plt.xlabel('State')
plt.ylabel('Average Total Day Calls')
plt.title('Scatter Plot: State vs Average Day Calls')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Boxplot of Total day calls
plt.figure(figsize=(8,6))
plt.boxplot(df['Total day calls'])
plt.title('Boxplot of Total Day Calls')
plt.ylabel('Total Day Calls')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.boxplot(x='Churn', y='Total day calls', data=df)
plt.title('Total Day Calls by Churn Status')
plt.xlabel('Churn')
plt.ylabel('Total Day Calls')
plt.show()
```
